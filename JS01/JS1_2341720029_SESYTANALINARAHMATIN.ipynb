1. PyPREP
•	Menerapkan re-referencing rata-rata (average reference) untuk meningkatkan kualitas sinyal.
•	Didesain untuk EEG, tetapi fungsi seperti deteksi outlier kanal dan penghapusan noise dapat diterapkan pada data multi-kanal lain yang memiliki karakteristik mirip (misalnya biosignal lain), jadi tidak sepenuhnya terbatas hanya EEG.
•	Menghilangkan atau menurunkan gangguan arus listrik 50/60 Hz melalui notch filtering.
•	Mengidentifikasi dan menangani kanal yang rusak menggunakan algoritme seperti robust referencing.
•	Membersihkan sinyal multi-kanal dari noise dengan metode otomatis (pipeline PREP).
2. SciPy
•	Pemrosesan citra dasar: konvolusi, filter, morfologi (di modul scipy.ndimage).
•	Fungsi jarak dan klastering (di modul scipy.spatial dan scipy.cluster).
•	Analisis statistik dan probabilitas: uji hipotesis, distribusi, regresi.
•	Menyediakan fungsi integrasi numerik, solusi persamaan diferensial, optimasi, dan interpolasi data.
•	Sering digunakan untuk komputasi ilmiah dan teknik karena memiliki banyak submodul fungsional.
•	Analisis sinyal: filter digital/analog, transformasi Fourier (FFT), spektrogram.
3. Weights & Biases (wandb)
•	Memantau serta menampilkan grafik metrik performa model machine learning (seperti loss, akurasi, F1 score, dll.) secara langsung.
•	Menyimpan detail konfigurasi eksperimen dan parameter model sehingga hasil dapat diulang kembali.
•	Menyediakan tampilan dashboard interaktif untuk melakukan perbandingan antar percobaan atau training model.
•	Mendukung penyimpanan dan pengelolaan artefak eksperimen (model terlatih, dataset, dsb.) agar lebih terorganisir.
•	Terintegrasi dengan berbagai framework ML populer (misalnya TensorFlow, PyTorch, Scikit-Learn) sehingga proses logging berjalan otomatis.
•	Memungkinkan kolaborasi dalam tim dengan sistem akses berdasarkan akun maupun proyek.
4. pyECG
•	Mendukung pembacaan serta pengolahan sinyal elektrokardiogram (ECG) dari beragam format file, seperti WFDB atau ISHNE.
•	Melakukan identifikasi kompleks QRS, interval PR/QT, serta karakteristik penting lain dari sinyal ECG.
•	Menyediakan analisis terkait denyut jantung (heart rate) dan variabilitasnya (HRV).
•	Menampilkan sinyal ECG dalam bentuk visual untuk kebutuhan analisis manual maupun berbasis algoritma.
•	Bermanfaat untuk pengujian algoritma pendeteksi aritmia serta studi klinis di bidang kardiologi
